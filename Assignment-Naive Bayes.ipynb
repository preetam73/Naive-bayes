```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
string_columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
for i in string_columns:
    data_train[i]=label_encoder.fit_transform(data_train[i])
    data_test[i]=label_encoder.fit_transform(data_test[i])
```

```python
col_name = list(data_train.columns)
```

```python
col_name
```

```python
x_train = data_train[col_name[0:13]]
x_train
```

```python
y_train = data_train[col_name[13]]
y_train
```

```python
x_test = data_test[col_name[0:13]]
x_test
```

```python
y_test = data_test[col_name[13]]
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
Gnbmodel=GaussianNB()
```

```python
train_pred_gau=Gnbmodel.fit(x_train,y_train).predict(x_train)
train_pred_gau
```

```python
test_pred_gau=Gnbmodel.fit(x_train,y_train).predict(x_test)
test_pred_gau
```

```python
train_acc_gau=np.mean(train_pred_gau==y_train)
```

```python
train_acc_gau
```

```python
test_acc_gau=np.mean(test_pred_gau==y_test)
```

```python
test_acc_gau
```

```python
from sklearn.naive_bayes import MultinomialNB
Mnbmodel=MultinomialNB()
```

```python
train_pred_multi=Mnbmodel.fit(x_train,y_train).predict(x_train)
train_pred_multi
```

```python
test_pred_multi=Mnbmodel.fit(x_train,y_train).predict(x_test)
test_pred_multi
```

```python
train_acc_multi=np.mean(train_pred_multi==y_train)
train_acc_multi
```

```python
test_pred_multi=np.mean(test_pred_multi==y_test)
test_pred_multi
```
